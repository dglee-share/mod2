```python
import re

names=[]
with open('name_gender_dataset.csv', 'r', encoding='utf8') as f:
    raw = f.readlines()

for line in raw[1:]:
    names.append(line.split(',')[0])
    
#rint(names)
```

```python
list1 = ['abcd','defg']

for x in list1:
    m = re.search('^.b.{2,5}$', x)
    if m is not None:
        print(m.group(0))
```

```python
def test_re(text_list,re_str):
    for x in text_list:
        m = re.search(re_str, x)
        if m is not None:
            print(m.group(0))
```
